```python
import pandas as pd

ratings = pd.read_csv('data/ratings_movies.csv', sep=',')
ratings
```

```python
import re 
def get_year_release(arg):
    #находим все слова по шаблону "(DDDD)"
    candidates = re.findall(r'\(\d{4}\)', arg) 
    # проверяем число вхождений
    if len(candidates) > 0:
        #если число вхождений больше 0,
	#очищаем строку от знаков "(" и ")"
        year = candidates[0].replace('(', '')
        year = year.replace(')', '')
        return int(year)
    else:
        #если год не указан, возвращаем None
        return None
ratings['year_release'] = ratings['title'].apply(get_year_release)
ratings['year_release'].info()

```

```python
ratings[ratings['year_release'] == 1999.0].min()
```

```python
ratings['genres'].nunique()
movies_2010 = ratings[ratings['year_release'] == 2010.0]
movies_2010_grouped = movies_2010.groupby(by='genres').mean(numeric_only=True).sort_values(by='rating')
movies_2010_grouped
```

```python
#шаг 1 - выцепляем год выставления рейтинга из таблицы
ratings['date'] = pd.to_datetime(ratings['date'], dayfirst=False)
ratings['year_rating'] = ratings['date'].dt.year
#шаг 2 - отвечаем на вопросы
```

Для жанра Comedy прослеживается тенденция падения рейтинга с каждым годом (с 1996 по 2018).
